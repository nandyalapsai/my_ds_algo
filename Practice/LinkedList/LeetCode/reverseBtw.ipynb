```python
class Node:
    def __init__(self,value):
        self.value = value
        self.next = None
class LinkedList:
    def __init__(self, value):
        new_node = Node(value)
        self.head = new_node
        self.length = 1

    def append(self, value):
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node
        self.length += 1
        return True
    
    def print_list(self):
        temp = self.head
        while temp:
            print(f"{temp.value}->", end="")
            temp = temp.next
        print("None")

    def reverse_btw(self,start_index,end_index):
        current = self.head
        left_end_node = current
        right_end_node = None
        start_index_node = None
        last_rev_node = None
        prev = None

        if start_index >= self.length or end_index >= self.length:
            return
        if self.length<=1:
            return

        for i in range(self.length):
            print(current.value)
            print(left_end_node.value)
            print("===")
            if current:
                if i>= start_index and i<=end_index:
                    if i==start_index:
                        start_index_node = current
                        prev = current
                        current = current.next
                        prev.next=None
                        last_rev_node = start_index_node
                    elif i==end_index:
                        if start_index == 0 and end_index == self.length -1:
                            left_end_node.next = None
                            prev = current
                            current = current.next
                            prev.next = last_rev_node
                            self.head = prev
                        else:
                            prev = current
                            current = current.next
                            prev.next = last_rev_node
                            left_end_node.next = prev
                            right_end_node = current
                            start_index_node.next = right_end_node
                    else:
                        prev = current
                        current = current.next
                        prev.next = last_rev_node
                        last_rev_node = prev
                else:
                    left_end_node = current
                    current = current.next

        def reverse_btw(self, start_index, end_index):
            # Validate inputs
            if start_index < 0 or end_index >= self.length or start_index > end_index:
                return
            if self.length <= 1 or start_index == end_index:
                return
            
            # Find the node before start_index (left_prev)
            left_prev = None
            current = self.head
            
            # Navigate to the start position
            for i in range(start_index):
                left_prev = current
                current = current.next
            
            # Store the first node to be reversed
            reverse_start = current
            
            # Reverse the sublist from start_index to end_index
            prev = None
            for i in range(end_index - start_index + 1):
                next_temp = current.next
                current.next = prev
                prev = current
                current = next_temp
            
            # Connect the reversed sublist back to the original list
            # prev is now the new head of the reversed sublist
            # current is the node after end_index
            
            if left_prev:  # start_index > 0
                left_prev.next = prev
            else:  # start_index == 0, update head
                self.head = prev
            
            # Connect the end of reversed sublist to the rest
            reverse_start.next = current
                    
```
